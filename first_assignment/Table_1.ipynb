```python
import pandas as pd
import numpy as np
```

```python
socio_eco = pd.read_csv(r'a-fanning-doughnut-v3-a0460e5\Analysis-Final\myData\2_20250108_doughnutv3-data.csv')
```

```python
socio_eco.head()
```

```python
socio = socio_eco[socio_eco['domain'] == 'social']
```

```python
socio.head()
```

```python
socio.reset_index(drop=True)
```

```python
socio = socio[socio['type'] != 'national aggregate']
```

```python
# drop unnecessary columns
socio = socio.drop(columns=['type', 'group', 'populationTotal', 'GNIperCap', 'domain'])
```

```python
# see state of missing values
socio.isna().sum()
```

```python
socio = socio[socio['value'].notna()]
```

```python
socio_ind = (
    socio.sort_values('date')
       .groupby(['dimension', 'indicator'])
       .agg({
           'date': ['first', 'last'],
           'value': ['first', 'last']
       })
)
```

```python
socio_ind = socio_ind.round(2)
socio_ind = socio_ind.reset_index()
socio_ind
```

```python
table_1 = socio_ind[socio_ind['indicator'] != 'EUshare_hzdHealth'].copy()
table_1 = table_1.round(2) # use two decimals like in assignment sheet
table_1
```
